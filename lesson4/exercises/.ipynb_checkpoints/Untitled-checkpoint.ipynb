## E01: I did not get around to seeing what happens when you initialize all weights and biases to zero. Try this and train the neural net. You might think either that 1) the network trains just fine or 2) the network doesn't train at all, but actually it is 3) the network trains but only partially, and achieves a pretty bad final performance. Inspect the gradients and activations to figure out what is happening and why the network is only partially training, and what part is being trained exactly.

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt # for making figures
%matplotlib inline
```

```python
# read in all the words
words = open('../names.txt', 'r').read().splitlines()
words[:8]
```

```python
len(words)
```

```python
# build the vocabulary of characters and mappings to/from integers
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}
vocab_size = len(itos)
print(itos)
print(vocab_size)
```

```python
# build the dataset
block_size = 3 # context length: how many characters do we take to predict the next one?

def build_dataset(words):  
  X, Y = [], []
  
  for w in words:
    context = [0] * block_size
    for ch in w + '.':
      ix = stoi[ch]
      X.append(context)
      Y.append(ix)
      context = context[1:] + [ix] # crop and append

  X = torch.tensor(X)
  Y = torch.tensor(Y)
  print(X.shape, Y.shape)
  return X, Y

import random
random.seed(42)
random.shuffle(words)
n1 = int(0.8*len(words))
n2 = int(0.9*len(words))

Xtr,  Ytr  = build_dataset(words[:n1])     # 80%
Xdev, Ydev = build_dataset(words[n1:n2])   # 10%
Xte,  Yte  = build_dataset(words[n2:])     # 10%
```

```python
# MLP revisited
n_embd = 10 # the dimensionality of the character embedding vectors
n_hidden = 200 # the number of neurons in the hidden layer of the MLP

g = torch.Generator().manual_seed(2147483647) # for reproducibility
C  = torch.randn((vocab_size, n_embd),            generator=g)
W1 = torch.randn((n_embd * block_size, n_hidden), generator=g) * 0
b1 = torch.randn(n_hidden,                        generator=g) * 0
W2 = torch.randn((n_hidden, vocab_size),          generator=g) * 0
b2 = torch.randn(vocab_size,                      generator=g) * 0


parameters = [C, W1, b1, W2, b2]
print(sum(p.nelement() for p in parameters)) # number of parameters in total
for p in parameters:
  p.requires_grad = True
```

```python
max_steps = 200000
batch_size = 32
loss_history = []

for i in range(max_steps):
    # --- Minibatch Construction ---
    ix = torch.randint(0, Xtr.shape[0], (batch_size,), generator=g)
    Xb, Yb = Xtr[ix], Ytr[ix]
    
    # --- Forward Pass ---
    emb = C[Xb]                    # Embed the characters
    embcat = emb.view(emb.shape[0], -1)  # Concatenate embeddings for the context
    hpreact = embcat @ W1 + b1       # Pre-activation for the hidden layer
    h = torch.tanh(hpreact)          # Apply non-linearity (tanh)
    logits = h @ W2 + b2             # Compute output logits
    loss = F.cross_entropy(logits, Yb)
    
    # --- Backward Pass ---
    for p in parameters:
        p.grad = None
    loss.backward()
    
    # Optional: Inspect gradients or activations at the first iteration
    if i == 0:
        print("=== Initial Activation & Gradient Stats ===")
        print("hpreact: mean =", hpreact.mean().item(), ", std =", hpreact.std().item())
        print("h (after tanh): mean =", h.mean().item(), ", std =", h.std().item())
        print("W1 grad: mean =", W1.grad.mean().item(), ", std =", W1.grad.std().item())
        print("W2 grad: mean =", W2.grad.mean().item(), ", std =", W2.grad.std().item())
        print("b1 grad: mean =", b1.grad.mean().item(), ", std =", b1.grad.std().item())
        print("b2 grad: mean =", b2.grad.mean().item(), ", std =", b2.grad.std().item())
        print("C grad: mean =", C.grad.mean().item(), ", std =", C.grad.std().item())
    
    # --- Parameter Update ---
    lr = 0.1 if i < 100000 else 0.01  # Learning rate schedule
    for p in parameters:
        p.data += -lr * p.grad
    
    # --- Track and Print Loss ---
    loss_history.append(loss.log10().item())
    if i % 10000 == 0:
        print(f'{i:7d}/{max_steps:7d}: loss = {loss.item():.4f}')
```

```python
plt.plot(loss_history)
plt.xlabel('Iteration')
plt.ylabel('Log10(Loss)')
plt.title('Training Loss over Iterations')
plt.show()
```

```python
# Cell: Training loop with gradient tracking and plotting

# Reset parameter gradients and history lists
grad_norm_C = []
grad_norm_W1 = []
grad_norm_b1 = []
grad_norm_W2 = []
grad_norm_b2 = []
loss_history = []

# We'll run a shorter loop for visualization (you can adjust max_steps)
max_steps = 5000
batch_size = 32

for i in range(max_steps):
    # Create minibatch
    ix = torch.randint(0, Xtr.shape[0], (batch_size,), generator=g)
    Xb, Yb = Xtr[ix], Ytr[ix]
    
    # Forward pass
    emb = C[Xb]                    # Embedding lookup
    embcat = emb.view(emb.shape[0], -1)  # Flatten embeddings
    hpreact = embcat @ W1 + b1       # Hidden pre-activation
    h = torch.tanh(hpreact)          # Hidden activation (tanh)
    logits = h @ W2 + b2             # Output logits
    loss = F.cross_entropy(logits, Yb)
    
    # Zero gradients
    for p in parameters:
        p.grad = None
    
    # Backward pass
    loss.backward()
    
    # Record gradient norms (L2 norm) for each parameter
    grad_norm_C.append(C.grad.norm().item())
    grad_norm_W1.append(W1.grad.norm().item())
    grad_norm_b1.append(b1.grad.norm().item())
    grad_norm_W2.append(W2.grad.norm().item())
    grad_norm_b2.append(b2.grad.norm().item())
    
    # Update parameters
    lr = 0.1 if i < 100000 else 0.01  # Learning rate schedule (not relevant here since max_steps is small)
    for p in parameters:
        p.data += -lr * p.grad
    
    # Track loss for plotting
    loss_history.append(loss.item())

# Plot the gradient norms over iterations
plt.figure(figsize=(12, 6))
iterations = range(max_steps)
plt.plot(iterations, grad_norm_C, label="C (Embedding)", color='blue')
plt.plot(iterations, grad_norm_W1, label="W1 (Hidden weights)", color='orange')
plt.plot(iterations, grad_norm_b1, label="b1 (Hidden biases)", color='green')
plt.plot(iterations, grad_norm_W2, label="W2 (Output weights)", color='red')
plt.plot(iterations, grad_norm_b2, label="b2 (Output biases)", color='purple')
plt.xlabel('Iteration')
plt.ylabel('Gradient Norm (L2)')
plt.title('Gradient Norms of Parameters Over Training Iterations')
plt.legend()
plt.show()

```
